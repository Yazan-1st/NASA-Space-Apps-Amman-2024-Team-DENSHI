```python
# prompt: using pandas  open the temperatures_index.csv file

import pandas as pd

# Assuming the file is in the current working directory
df = pd.read_csv('/content/temperatures_index.csv')

# You can print the DataFrame to see its contents
df
```

```python
# @title Lowess(5)

from matplotlib import pyplot as plt
df['Lowess(5)'].plot(kind='line', figsize=(8, 4), title='Lowess(5)')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
df = df[df['Year'] >= 1993]
df
```

```python
# @title Lowess(5)

from matplotlib import pyplot as plt
grouped_df = df.groupby('Year')['Lowess(5)'].mean()
grouped_df
```

```python
grouped_df.plot(kind='line', figsize=(8, 4), title='Lowess(5)', marker='o')
```

```python
grouped_df.to_csv('temperatures.csv')
```
