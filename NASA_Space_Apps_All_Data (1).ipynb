```python
# prompt: open a csv file in padnas

import pandas as pd

df = pd.read_csv('final_dataset.csv')
df
```

```python
import plotly.express as px

# Create a list of column names to iterate through
columns_to_plot = ['co2_avg', 'methan_avg', 'temperatures_avg', 'GMSL No GIA']

# Loop through the columns and create an interactive line chart for each
for column in columns_to_plot:
    fig = px.line(df, x='Year', y=column, title=f'{column} over time', markers=True)
    fig.update_layout(xaxis_title='Year', yaxis_title=column)
    fig.show()
```
